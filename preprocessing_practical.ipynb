```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords
import re
import pandas as pd
```

```python
df = pd.read_csv(r"data\tripadvisor_hotel_reviews.csv")
pd.set_option("display.max_colwidth", None)
df.info()
```

```python
df.head()
```

```python
df["Review"][0]
```

```python
df["review_lowercase"] = df["Review"].str.lower()
df.head()
```

```python
# removing stopwords
en_stopwords = stopwords.words("english")
en_stopwords
```

```python
# removing not form the stopwords dictionary
en_stopwords.remove("not")
```

```python
df["review_no_stopwords"] = df["review_lowercase"].apply(lambda x: ' '.join([word for word in x.split() if word not in (en_stopwords)]))
```

```python
df[["review_lowercase", "review_no_stopwords"]].head()
```

 remove punctuation

```python
df["reveiw_no_stopwords_no_punct"] = df.apply(lambda x: re.sub(r"[*]", "star", x["review_no_stopwords"]), axis=1)
```

```python
df.head()
```

```python
df["reveiw_no_stopwords_no_punct"] = df.apply(lambda x: re.sub(r"[^\w\s]", "", x["reveiw_no_stopwords_no_punct"]), axis=1)
```

```python
df.head()
```

tokenization

```python
df["tokenized"] = df.apply(lambda x: word_tokenize(x["reveiw_no_stopwords_no_punct"]), axis=1)
df["tokenized"][0]
```

```python
ps = PorterStemmer()
df["stemmeized"] = df.apply(lambda x: [ps.stem(word)  for word in x['tokenized']], axis=1)
df[["tokenized", "stemmeized"]].head()
```

```python
lem = WordNetLemmatizer()
df["lemmentized"] = df.apply(lambda x: [lem.lemmatize(word) for word in x["tokenized"]], axis=1)
df[["tokenized", "stemmeized", "lemmentized"]].head()
```

```python
df.head()
```

```python
token_clean = sum(df["lemmentized"], [])
```

```python
unigrams = (pd.Series(nltk.ngrams(token_clean, 1)).value_counts())
print(unigrams)
```

```python
bigrams = (pd.Series(nltk.ngrams(token_clean, 2)).value_counts())
print(bigrams)
```

```python
trigrams = (pd.Series(nltk.ngrams(token_clean, 3)).value_counts())
print(trigrams)
```
