## Rule-Based Sentiment Analysis
is not a machine learning model but rather it uses a set of predefined rules that connect words to emotions or attitudes

Polarity score: is a numerical value that shows whether the word is postive or negative
scores usually ranges from [-1...0..1] where an number above zero indicate a postive sentiment

NB: Rule Base Sentiment can't handle more subtlr expression, such as sarcasm or irony

Python packages that follows the rule-based sentiment is the `TextBlob` and `VADER`

`Lexicon`: a collection of words paired with their predefined scores

```python
sentence_1 = "i had a great time at the movie it was really funny"
sentence_2 = "i had a great time at the movie but the parking was terrible"
sentence_3 = "i had a great time at the movie but the parking wasn't great"
sentence_4 = "i went to see a movie"
```

### <H3>TextBlob<H3>

```python
from textblob import TextBlob
```

```python
def textblob_polarity(sentence):
    sentiment_score = TextBlob(sentence)
    print(sentence, end=' ')
    print(sentiment_score.sentiment.polarity)
```

```python
textblob_polarity(sentence_1)
textblob_polarity(sentence_2)
textblob_polarity(sentence_3)
textblob_polarity(sentence_4)
```

### <h3>VADAR</h3>

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
vadar_sentiment = SentimentIntensityAnalyzer()
def vadar_polarity(sentence):
    print(sentence)
    print(vadar_sentiment.polarity_scores(sentence), end="\n\n")
```

```python
vadar_polarity(sentence_1)
vadar_polarity(sentence_2)
vadar_polarity(sentence_3)
vadar_polarity(sentence_4)
```

## Pre-trained Transformer
Transformeer are based on Deep learning anc can capture how words influence and relate to each other
even when they appear far apart in a sentence 
Transformer are crutail because they understand context


```python
import transformers
from transformers import pipeline
```

```python
sentiment_pipeline = pipeline("sentiment-analysis")
```

```python
def score(sentence):
    print(sentence, end=" ")
    sentiment_pipeline(sentence)
```

```python
score(sentence_1)
score(sentence_2)
score(sentence_3)
score(sentence_4)
```
