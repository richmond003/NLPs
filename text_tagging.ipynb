### Part Of Speach Tagging (POS)
Labeling each word in a sentence with its grammatical role

```python
import spacy
import pandas as pd
```

```python
text_data =  open(r"txt_files\pos_tagging.txt", "r")
```

```python
# load spacy model
nlp = spacy.load("en_core_web_sm") # is a an english core model that is light weight and faster
emma_ja = text_data.read()
print(emma_ja)
```

A spacy document: an object that stores the text along <br>
with all the linguistic information spaCy generates

```python
spacy_doc = nlp(emma_ja) # spaCy documents each word is already tokenized and tagged
spacy_doc
```

```python
pos_df = pd.DataFrame(columns=["token", "pos_tag"]) # empty dataframe
for token in spacy_doc:
    pos_df = pd.concat([pos_df,
                        pd.DataFrame.from_records([{'token': token.text, 'pos_tag': token.pos_}])],  ignore_index = True) # one roll table using pd.DataFrame.from_records
    # pd.concat joins the row in to the main dataframe
```

```python
pos_df.head(15)
```

```python
pos_df_counts = pos_df.groupby(["token", "pos_tag"]).size().reset_index(name="counts").sort_values(by="counts", ascending=False)
pos_df_counts.head(10)
```

```python
pos_df_poscounts = pos_df_counts.groupby(["pos_tag"])["token"].count().sort_values(ascending=False)
pos_df_poscounts.head(10)
```

```python
nouns = pos_df_counts[pos_df_counts.pos_tag == "NOUN"][:10]
nouns
```

### Named Entity Recognition (NER)
Is the process of finding and labeling essential pieces of information in text. <br>
NER uses a combination of rules and machine learning to dicover the entities

```python
from spacy import displacy # visiualization for spacy that alllows us to see entities highlighted directly in the text
from spacy import tokenizer
from IPython.display import HTML, display
import re
```

```python
nlp_ner = spacy.load("en_core_web_sm") # spacy model
```

```python
google_text = "Google was founded on September 4, 1998, by computer scientists Larry Page and Sergey Brin while they were PhD students at Stanford University in California. Together they own about 14% of its publicly listed shares and control 56% of its stockholder voting power through super-voting stock. The company went public via an initial public offering (IPO) in 2004. In 2015, Google was reorganized as a wholly owned subsidiary of Alphabet Inc. Google is Alphabet's largest subsidiary and is a holding company for Alphabet's internet properties and interests. Sundar Pichai was appointed CEO of Google on October 24, 2015, replacing Larry Page, who became the CEO of Alphabet. On December 3, 2019, Pichai also became the CEO of Alphabet."
```

```python
spacy_doc_ner = nlp_ner(google_text)
spacy_doc_ner
```

```python
for word in spacy_doc_ner.ents:
    print(f"{word} : {word.label_} {word.label}") # label with out the underscore (.lable) is a numeric ID od the entity
# while .label_ is a human-readable label
```

```python
def html_visiualizer(doc):
    html = displacy.render(doc, style="ent", jupyter=False)
    display(HTML(html))
```

```python
html_visiualizer(spacy_doc_ner)
```

```python
google_text_clean = re.sub(r"[^\w\s]", "", google_text).lower()
google_text_clean
```

```python
spacy_doc_clean = nlp_ner(google_text_clean)
```

```python
for word in spacy_doc_clean.ents:
    print(f"{word} : {word.label_}")
```

```python
html_visiualizer(spacy_doc_clean)
```
